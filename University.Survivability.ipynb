```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
univ = pd.read_csv('University.csv')
```

```python
univ.columns = univ.columns.str.replace(' ', '_')
univ['Average_Undergraduate_Tuition_and_Fees'] = univ['Average_Undergraduate_Tuition_and_Fees'].str.replace('$', '')
univ['Average_Undergraduate_Tuition_and_Fees'] = univ['Average_Undergraduate_Tuition_and_Fees'].str.replace(',', '')
univ['Average_Undergraduate_Tuition_and_Fees'] = univ['Average_Undergraduate_Tuition_and_Fees'].str.rstrip()
pd.to_numeric(univ['Average_Undergraduate_Tuition_and_Fees'])

univ['Full_Time_Enrollment'] = univ['Full_Time_Enrollment'].str.replace(',', '')
univ['Full_Time_Enrollment'] = pd.to_numeric(univ['Full_Time_Enrollment'])

univ['Average_Monthly_Search_Volume'] = univ['Average_Monthly_Search_Volume'].str.replace(',', '')
univ['Average_Monthly_Search_Volume'] = pd.to_numeric(univ['Average_Monthly_Search_Volume'])

univ['Endowment_per_Full_Time_Student'] = univ['Endowment_per_Full_Time_Student'].str.replace('$', '')
univ['Endowment_per_Full_Time_Student'] = univ['Endowment_per_Full_Time_Student'].str.replace(',', '')
univ['Endowment_per_Full_Time_Student'] = pd.to_numeric(univ['Endowment_per_Full_Time_Student'])


#univ = univ.sort_values(by =['Average_Undergraduate_Tuition_and_Fees'])

univ
```

```python


def label_states (row):
    if row['State'] in ['ME', 'NH', 'VT', 'MA', 'RI', 'CT', 'NY', 'PA', 'NJ']:
        return 'north-east'
    if row['State'] in ['WI', 'MI', 'IL', 'IN', 'OH', 'ND', 'SD', 'NE', 'KS', 'MN', 'IA', 'MO']:
        return 'midwest'
    if row['State'] in ['DE', 'MM', 'DC', 'VA', 'WV', 'NC', 'SC', 'GA', 'FL', 'KY', 'TN', 'MS', 'AL', 'OK', 'TX', 'AR', 'LA']:
        return 'south'
    if row['State'] in ['MT', 'ID', 'WY', 'CO', 'NM', 'AZ', 'UT', 'NV', 'CA', 'OR', 'WA', 'AK',  'HI']:
        return 'west'
    
univ['region'] = univ.apply(lambda row: label_states(row), axis=1)
univ

```

```python


statscounts = univ.groupby(["Prof_G_Categorization", 'region'])['Value-to-Cost_Ratio']
statscounts.mean()
sc = statscounts.mean()
sc = sc.to_frame()
sc

```
